```python
import matplotlib.pyplot as plt

def differential_manchester_encode(data):
    encoded_data = ''
    last_bit = '0'

    for bit in data:
        if bit == '0':
            if last_bit == '0':
                encoded_data += '01'
            else:
                encoded_data += '10'
                last_bit = '0'
        elif bit == '1':
            if last_bit == '1':
                encoded_data += '01'
            else:
                encoded_data += '10'
                last_bit = '1'
        else:
            raise ValueError("0 혹은 1로 이루어진 값만 입력하시오.")

    return encoded_data

def plot_differential_manchester_encoding(data, encoded_data):
    fig, ax = plt.subplots(figsize=(8, 2))

    x = [i for i in range(len(encoded_data) * 2)]
    y = []

    for bit in encoded_data:
        if bit == '0':
            y.extend([1, 0])
        elif bit == '1':
            y.extend([0, 1])

    ax.step(x, y, where='post', color='blue')
    ax.set_title("Differential Manchester Encoding")
    ax.set_xlabel("비트 위치")
    ax.set_ylabel("전압 레벨 (0 또는 1)")

    ax.set_xticks(x)
    ax.set_xticklabels(list(data))
    ax.grid()

    plt.show()

try:
    user_input = int(input("임의의 숫자를 입력하세요: "))
    binary_data = bin(user_input)[2:]
    encoded_data = differential_manchester_encode(binary_data)

    print("차등 맨체스터 인코딩 결과:", encoded_data)

    plot_differential_manchester_encoding(binary_data, encoded_data)

except Exception as e:
    print("오류 발생:", e)
```
